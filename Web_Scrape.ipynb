```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://en.wikipedia.org/w/api.php"
params = {
    "action": "query",
    "prop": "revisions",
    "rvprop": "content",
    "rvparse": "",
    "format": "json",
    "titles": "List_of_defunct_airlines_of_Nigeria"
}

response = requests.get(url, params=params)
if response.status_code == 200:
    data = response.json()
    page = next(iter(data['query']['pages'].values()))
    html_content = page['revisions'][0]['*']

    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the specific table
    table = soup.find('table', class_='wikitable')
    if table:
        rows_list = []
        for row in table.find_all('tr')[1:]:  # Skip the header row
            cols = row.find_all('td')
            if len(cols) > 5:  # Ensure there are enough columns
                airline_name = cols[0].text.strip()
                iata = cols[1].text.strip()
                icao = cols[2].text.strip()
                callsign = cols[3].text.strip()
                commenced = cols[4].text.strip()
                ceased = cols[5].text.strip()
                notes = cols[6].text.strip() if len(cols) > 6 else ""
                rows_list.append([airline_name, iata, icao, callsign, commenced, ceased, notes])

        df2 = pd.DataFrame(rows_list, columns=['Airline Name', 'IATA', 'ICAO', 'Callsign', 'Commenced Operations', 'Ceased Operations', 'Notes'])
        print(df2)
    else:
        print("No table found.")
else:
    print("Failed to fetch data. Status code:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://en.wikipedia.org/wiki/List_of_airports_in_Nigeria"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the specific table
table = soup.find('table', {'class': 'wikitable sortable'})

# Extracting data from the table
rows = table.find_all('tr')
data = []
for row in rows[1:]:
    cols = row.find_all('td')
    if cols:
        city_served = cols[0].text.strip()
        state = cols[1].text.strip()
        icao = cols[2].text.strip()
        iata = cols[3].text.strip()
        airport_name = cols[4].text.strip()
        data.append([city_served, state, icao, iata, airport_name])

# Creating DataFrame
df3 = pd.DataFrame(data, columns=['City Served', 'State', 'ICAO', 'IATA', 'Airport Name'])
print(df3)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# List of airlines with their Wikipedia suffixes
airlines = [
    "Aero_Contractors_(Nigeria)",
    "Air_Peace",
    "Allied_Air",
    "Arik_Air",
    "Azman_Air",
    "Dana_Air",
    "Dornier_Aviation_Nigeria",
    "Green_Africa_Airways",
    "Ibom_Air",
    "K-Impex_Airline",
    "Kabo_Air",
    "Max_Air",
    "Overland_Airways",
    "TAT_Nigeria",
    "United_Nigeria_Airlines",
    "ValueJet_(Nigeria)",
    "West_Link_Airlines",
    "XEJet"
]

def fetch_fleet_info(airline_suffix):
    url = f"https://en.wikipedia.org/wiki/{airline_suffix}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all 'wikitable' and target the last one on the page
    tables = soup.find_all('table', {'class': 'wikitable'})
    if tables:
        table = tables[-1]  # Select the last table
        try:
            df = pd.read_html(str(table), flavor='bs4')[0]

            # Flatten MultiIndex columns (if any)
            if isinstance(df.columns, pd.MultiIndex):
                df.columns = [' '.join(col).strip() for col in df.columns.values]
            else:
                df.columns = df.columns.str.strip()  # Clean single-level columns

            # Normalize column names to handle variations
            df.columns = df.columns.str.lower()  # Convert column names to lowercase

            # Check for variations of the required columns
            aircraft_col = None
            in_service_col = None
            orders_col = None

            for col in df.columns:
                if 'aircraft' in col:
                    aircraft_col = col
                elif 'in service' in col or 'in-service' in col or 'in fleet' in col or 'active' in col or 'total' in col:
                    in_service_col = col
                elif 'orders' in col or 'on order' in col or 'order' in col:
                    orders_col = col

            # If required columns are found, rearrange them
            if aircraft_col and in_service_col:
                # Include 'Orders' if available, otherwise set to None
                if orders_col:
                    df = df[[aircraft_col, in_service_col, orders_col]]
                    df.columns = ['Aircraft', 'In service', 'Orders']
                else:
                    df = df[[aircraft_col, in_service_col]]
                    df['Orders'] = None  # Add missing column
                    df.columns = ['Aircraft', 'In service', 'Orders']

                # Convert "In service" and "Orders" to integers (if possible)
                df['In service'] = pd.to_numeric(df['In service'], errors='coerce').fillna(0).astype(int)
                if 'Orders' in df.columns:
                    df['Orders'] = pd.to_numeric(df['Orders'], errors='coerce').fillna(0).astype(int)

                # Filter out rows where "Aircraft" contains "Total"
                df = df[~df['Aircraft'].str.contains('Total', case=False, na=False)]

                return df
            else:
                return None  # Return None if required columns are missing
        except ValueError as e:
            return None  # Return None if there's an error reading the table
    return None  # Return None if no table is found

# List to hold all fleet data
all_fleet_data = []

# Iterate over the airline list and fetch fleet data
for airline in airlines:
    fleet_info = fetch_fleet_info(airline)
    airline_name = airline.replace('_', ' ').replace("(Nigeria)", "").strip()

    if fleet_info is not None:
        # Add the airline name as the first column
        fleet_info.insert(0, 'Airline', airline_name)
        all_fleet_data.append(fleet_info)
    else:
        print(f"No fleet information found for {airline_name}.")

# Combine all fleet data into a single DataFrame
if all_fleet_data:
    combined_df = pd.concat(all_fleet_data, ignore_index=True)
    print(combined_df.to_string(index=False))
else:
    print("No fleet information found for any airline.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# List of airlines with their Wikipedia suffixes
airlines = [
    "Aero_Contractors_(Nigeria)",
    "Air_Peace",
    "Allied_Air",
    "Arik_Air",
    "Azman_Air",
    "Dana_Air",
    "Dornier_Aviation_Nigeria",
    "Green_Africa_Airways",
    "Ibom_Air",
    "K-Impex_Airline",
    "Kabo_Air",
    "Max_Air",
    "Overland_Airways",
    "TAT_Nigeria",
    "United_Nigeria_Airlines",
    "ValueJet_(Nigeria)",
    "West_Link_Airlines",
    "XEJet"
]

def fetch_destinations(airline_suffix):
    url = f"https://en.wikipedia.org/wiki/{airline_suffix}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the "Destinations" section
    destinations_section = soup.find('span', {'id': 'Destinations'})
    if not destinations_section:
        # Try alternative section titles
        destinations_section = soup.find('span', {'id': 'Fleet_and_destinations'})
        if not destinations_section:
            return []

    # Find the next sibling element (usually a list, table, or paragraph)
    destinations_content = destinations_section.find_next()
    destinations = []

    # Case 1: List format (bulleted list with en dash separator)
    if destinations_content.name == 'ul':
        for li in destinations_content.find_all('li'):
            text = li.get_text(strip=True)
            if '–' in text:  # Handle en dash separator
                city, airport = text.split('–', 1)
                destinations.append({
                    'Country': 'Nigeria',  # Assume domestic flights
                    'City': city.strip(),
                    'Airport': airport.strip(),
                    'Notes': ''
                })
            else:
                destinations.append({
                    'Country': 'Nigeria',  # Assume domestic flights
                    'City': text,
                    'Airport': '',
                    'Notes': ''
                })

    # Case 2: Table format
    elif destinations_content.name == 'table':
        headers = [th.get_text(strip=True) for th in destinations_content.find_all('th')]
        for row in destinations_content.find_all('tr')[1:]:  # Skip header row
            cols = row.find_all('td')
            if len(cols) >= len(headers):  # Ensure there are enough columns
                destination = {}
                for i, header in enumerate(headers):
                    destination[header] = cols[i].get_text(strip=True)
                # If 'Country' is missing, assume it's Nigeria (domestic flights)
                if 'Country' not in destination or not destination['Country']:
                    destination['Country'] = 'Nigeria'
                destinations.append(destination)

    # Case 3: Paragraph format (not handled in this example)
    # You can add logic here if needed.

    return destinations

# List to hold all airline data
all_airline_data = []

# Iterate over the airline list and fetch destination data
for airline in airlines:
    destinations_info = fetch_destinations(airline)
    airline_name = airline.replace('_', ' ').replace("(Nigeria)", "").strip()

    if destinations_info:  # If destinations are successfully fetched
        for destination in destinations_info:
            # Add the airline name and destination details to the list
            all_airline_data.append({
                'Airline': airline_name,
                'Country': destination.get('Country', 'Nigeria'),  # Default to Nigeria
                'City': destination.get('City', ''),
                'Airport': destination.get('Airport', ''),
                'Notes': destination.get('Notes', '')
            })
    else:
        # If no destinations are found, add a row with the airline name and no destination details
        all_airline_data.append({
            'Airline': airline_name,
            'Country': '',
            'City': '',
            'Airport': 'No destinations found',
            'Notes': ''
        })

# Convert the list to a DataFrame for better visualization
df = pd.DataFrame(all_airline_data)

# Reorder columns to have 'Airline' first
df = df[['Airline', 'Country', 'City', 'Airport', 'Notes']]

# Save the DataFrame to a CSV file (optional)
df.to_csv('airline_destinations.csv', index=False)

# Print the DataFrame
print(df.to_string(index=False))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def fetch_aero_second_destinations():
    url = "https://en.wikipedia.org/wiki/Aero_Contractors_(Nigeria)"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the 'Destinations' section, expected to be near an <ol>
    section = soup.find('span', id='Destinations')
    if section:
        ol_elements = section.find_all_next('ol')  # Find all ordered lists following the 'Destinations' tag
        if len(ol_elements) > 1:  # Ensure there is at least a second ordered list
            ol = ol_elements[1]  # Target the second ordered list
            data = []
            list_items = ol.find_all('li')
            for item in list_items:
                text = item.get_text(strip=True)
                parts = text.split(' - ')  # Split by " - " as specified
                if len(parts) == 2:
                    city, airport = parts
                    data.append({'Country': 'Nigeria', 'City': city.strip(), 'Airport': airport.strip()})
                else:
                    # In case the format is not as expected, handle gracefully
                    data.append({'Country': 'Nigeria', 'City': text, 'Airport': 'Unknown'})

            return data
    return "No second ordered list found under Destinations."

# Fetch Aero Contractors' second list of destinations
aero_second_destinations = fetch_aero_second_destinations()

# Check if data was found or a string message was returned
if isinstance(aero_second_destinations, list):
    df = pd.DataFrame(aero_second_destinations)
    print(df)
else:
    print(aero_second_destinations)
```
